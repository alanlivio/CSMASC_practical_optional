```python
import boto3
import json

MODEL_ID = "amazon.nova-lite-v1:0"
REGION = "eu-north-1"

client = boto3.client("bedrock-runtime", region_name=REGION)


def invoke(prompt: str) -> dict:
    response = client.invoke_model(
        modelId=MODEL_ID,
        body=json.dumps(
            {
                "messages": [
                    {
                        "role": "user",
                        "content": [{"text": prompt}],
                    }
                ]
            }
        ),
    )
    return json.loads(response["body"].read())
```
