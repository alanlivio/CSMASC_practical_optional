## Shared Helpers

```python
import boto3
import json

MODEL_ID = "amazon.nova-lite-v1:0"
REGION = "eu-north-1"

client = boto3.client("bedrock-runtime", region_name=REGION)


def invoke(prompt: str) -> dict:
    response = client.invoke_model(
        modelId=MODEL_ID,
        body=json.dumps(
            {
                "messages": [
                    {
                        "role": "user",
                        "content": [{"text": prompt}],
                    }
                ]
            }
        ),
    )
    return json.loads(response["body"].read())
```

## Task 1 Solution

```python
zero_shot_prompt = """Sulfuric acid reacts with sodium chloride,
and gives <chemical1>_____</chemical1> and <chemical2>_____</chemical2>:
The chemical1 and chemical 2 are:
"""

few_shot_prompt = """Tweet: \"I hate it when my phone battery dies.\": Sentiment: Negative
Tweet: \"My day has been great\": Sentiment: Positive
Tweet: \"This is the link to the article\": Sentiment: Neutral
Tweet: \"This new music video was incredible\" Sentiment:
"""
```

```python
zero_shot_result = invoke(zero_shot_prompt)
print(json.dumps(zero_shot_result, indent=2))
```

```python
few_shot_result = invoke(few_shot_prompt)
print(json.dumps(few_shot_result, indent=2))
```

## Task 2 Solution

```python
prompt_template = """
You are a helpful travel assistant.
Plan a trip to {destination} for {audience}.
Use a {tone} tone.
Write an enthusiastic headline and a short summary itinerary.
"""

test_cases = [
    {
        "destination": "Hanoi",
        "audience": "university students",
        "tone": "enthusiastic",
    },
    {
        "destination": "Hanoi",
        "audience": "retired travellers",
        "tone": "enthusiastic",
    },
    {
        "destination": "Hanoi",
        "audience": "university students",
        "tone": "formal",
    },
]
```

```python
results = []

for case in test_cases:
    prompt = prompt_template.format(**case)
    response_body = invoke(prompt)
    results.append({"case": case, "response": response_body})
    print(json.dumps(case, indent=2))
    print(json.dumps(response_body, indent=2))
    print("-" * 40)
```
